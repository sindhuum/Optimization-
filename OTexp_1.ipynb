```python
# Experiment No: 1
# Single Variable Optimization Problem using First Derivative Test
import sympy as sp
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Step 1: Define the variable and function
x = sp.Symbol('x')
f = x**3 - 6*x**2 + 9*x + 15
print("Function f(x):", f)
```

```python
# Step 2: First Derivative
f_prime = sp.diff(f, x)
print("First Derivative f'(x):", f_prime)
```

```python
# Step 3: Critical Points (solve f'(x)=0)
critical_points = sp.solve(f_prime, x)
print("Critical Points:", critical_points)
```

```python
# Step 4: First Derivative Test
for cp in critical_points:
    # Pick test points around cp
    left = cp - 1
    right = cp + 1
    f_left = f_prime.subs(x, left)
    f_right = f_prime.subs(x, right)

    if f_left > 0 and f_right < 0:
        nature = "Local Maximum"
    elif f_left < 0 and f_right > 0:
        nature = "Local Minimum"
    else:
        nature = "No extremum"

    print(f"At x = {cp}, f(x) = {f.subs(x, cp)}, Nature: {nature}")
```

```python
# Step 5: Plot the function with critical points
f_lambdified = sp.lambdify(x, f, 'numpy')
X = np.linspace(-1, 5, 400)
Y = f_lambdified(X)

plt.figure(figsize=(8,5))
plt.plot(X, Y, label="f(x) = x^3 - 6x^2 + 9x + 15", color="blue")
```

```python
# Plot critical points
for cp in critical_points:
    plt.scatter(float(cp), float(f.subs(x, cp)), color="red", zorder=5)
    plt.text(float(cp), float(f.subs(x, cp))+0.5,
             f"({float(cp)}, {float(f.subs(x, cp))})",
             ha='center', color="red")

plt.axhline(y=0, color="black", linewidth=0.5)
plt.title("Single Variable Optimization using First Derivative Test")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
